# **Processing Data for news -1.csv  ( df_news1 )**


```python
datafolder='/content/drive/MyDrive/Community Help /category classification/Data/'

df_news1 = pd.read_csv(datafolder+"news -1.csv")
```

```python
def Category_class(Category_int):
    if Category_int == 1:
        return 'Sports'
    if Category_int == 2:
        return 'Politics'
    if Category_int == 3:
        return 'TV'
    if Category_int == 4:
        return 'Intnews'
    if Category_int == 5:
        return 'Tech'
    if Category_int == 6:
        return 'Business'
    if Category_int == 7:
        return 'Health'
    else:
      return 'No Category'
```

```python
with open(datafolder+'train.txt') as f:
    lines = f.readlines()
```

```python
Modifiedlines=[]
for line in lines:
  li=line.split('|')
  Modifiedlines.append(li)
```

```python
len(Modifiedlines)
```

```python
for seq in range(len(Modifiedlines)):
    if len(Modifiedlines[seq])>2:
      Modifiedlines[seq][2]=Modifiedlines[seq][2].replace('|',"")
      mergestring=Modifiedlines[seq][1]+Modifiedlines[seq][2]
      Modifiedlines[seq][1]=mergestring
```

```python
def CategoryC1_assign(Category_str):
    if Category_str == "LEFT-NEWS" or Category_str == "POLITICS":
        return 'POLITICS'
    elif Category_str == "BUSINESS":
        return 'BUSINESS'
    elif Category_str == "ENTERTAINMENT" or Category_str == "COMEDY":
        return 'ENTERTAINMENT'
    elif Category_str == "SCIENCE & TECHNOLOGY" or Category_str == "TECHNOLOGY":
        return 'SCIENCE & TECHNOLOGY'
    elif Category_str == "SPORTS" or Category_str == "WORLD NEWS" or Category_str == "HEALTH" or Category_str == "HEALTHY LIVING" :
       return 'GENERAL-1'
    else:
      return Category_str
```

```python
def CategoryC2_assign(Category_str):
    if Category_str == "SPORTS":
        return 'SPORTS'
    elif Category_str == "WORLD NEWS":
        return 'WORLD NEWS'
    elif Category_str == "HEALTH" or Category_str == "HEALTHY LIVING":
        return 'HEALTH'
    elif Category_str == "ARTS & CULTURE" or Category_str == "TRAVEL" or  Category_str == "EDUCATION" or   Category_str == "CRIME"  or  Category_str == "RELIGION"  or  Category_str == "GREEN"  or Category_str == "WOMEN" or Category_str == "STYLE" or Category_str == "TASTE":
        return 'GENERAL-2'
    else:
      return Category_str
```

```python
Model1_columns=['POLITICS','BUSINESS','ENTERTAINMENT','SCIENCE & TECHNOLOGY','GENERAL-1']
Model2_columns=['SPORTS','WORLD NEWS','HEALTH','GENERAL-2']
Model2_columns=['ARTS & CULTURE','TRAVEL','EDUCATION','CRIME','RELIGION','GREEN','WOMEN','STYLE','TASTE']
```

```python
df_category_model1=df_category_unique1
df_category_model1["CATEGORY_CLASS_NEW"]=df_category_model1["CATEGORY_CLASS"].apply(CategoryC1_assign)
df_category_model1=df_category_model1[df_category_model1['CATEGORY_CLASS_NEW'].isin(Model1_columns)].reset_index()
df_category_model1.drop('index', axis='columns', inplace=True)
print(df_category_model1.shape)
df_category_model1.groupby(by=["CATEGORY_CLASS_NEW"]).count()
```

```python
df_category_model2=df_category_unique1
df_category_model2["CATEGORY_CLASS_NEW"]=df_category_model2["CATEGORY_CLASS"].apply(CategoryC2_assign)
df_category_model2=df_category_model2[df_category_model2['CATEGORY_CLASS_NEW'].isin(['SPORTS','WORLD NEWS','HEALTH','GENERAL-2'])].reset_index()
df_category_model2.drop('index', axis='columns', inplace=True)
print(df_category_model2.shape)
df_category_model2.groupby(by=["CATEGORY_CLASS_NEW"]).count()
```

```python
df_category_model3=df_category_unique1
df_category_model3["CATEGORY_CLASS_NEW"]=df_category_model3["CATEGORY_CLASS"]
df_category_model3=df_category_model3[df_category_model3['CATEGORY_CLASS_NEW'].isin(['ARTS & CULTURE','TRAVEL','EDUCATION','CRIME','RELIGION','GREEN','WOMEN','STYLE','TASTE'])].reset_index()
df_category_model3.drop('index', axis='columns', inplace=True)
print(df_category_model3.shape)
df_category_model3.groupby(by=["CATEGORY_CLASS_NEW"]).count()
```
