```python
def Category_class(Category_int):
    if Category_int == 0:
        return 'Neutral'
    if Category_int == 1:
        return 'Happy'
    if Category_int == 2:
        return 'Sad'
    if Category_int == 3:
        return 'Love'
    if Category_int == 4:
        return 'Anger'
```

```python
df_Emotion3=pd.DataFrame(Modifiedlines,columns=["CATEGORY_CLASS2","TITLE"])
```

```python
def Category_classmapping(Category_int):
    li=Category_int.split('.')
    if li[0].replace("[","").strip() != "0":
        return 'JOY'
    if li[1].replace("[","").strip() != "0":
        return 'FEAR'
    if li[2].replace("[","").strip() != "0":
        return 'ANGER'
    if li[3].replace("[","").strip() != "0":
        return 'SADNESS'
    if li[4].replace("[","").strip() != "0":
        return 'DISGUST'
    if li[5].replace("[","").strip() != "0":
        return 'SHAME'
    if li[6].replace("[","").strip() != "0":
        return 'GUILT'
```

```python
df_Emotion3["CATEGORY_CLASS"]=df_Emotion3["CATEGORY_CLASS2"].apply(Category_classmapping)
```

```python
def Category_classmapping1(Category_str):
   if Category_str == "ANGER" or Category_str == "DISGUST" or Category_str == "HATE":
        return 'ANGER'
   if Category_str == "FEAR" or Category_str == "WORRY":
        return 'NEUTRAL'
   if Category_str == "HAPPY" or Category_str == "HAPPINESS" or Category_str == "JOY" :
        return 'HAPPY'
   if Category_str == "LOVE" or Category_str == "FUN" or Category_str == "SURPRISE":
        return 'LOVE'
   if Category_str == "NEUTRAL" or Category_str == "EMPTY" or Category_str == "ENTHUSIASM" or Category_str == "GUILT" or Category_str == "NEUTRAL" or Category_str == "RELIEF" or Category_str == "SHAME":
        return 'NEUTRAL'
   if Category_str == "SAD" or Category_str == "SADNESS" :
        return 'SAD'
```

```python
df_category_Emotion_unique1["CATEGORY_CLASS"]=df_category_Emotion_unique1["CATEGORY_CLASS"].apply(Category_classmapping1)
```
